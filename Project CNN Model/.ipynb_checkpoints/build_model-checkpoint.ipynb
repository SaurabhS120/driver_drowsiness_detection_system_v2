# Building a model for predictiong open or close eye

## Load images into main memory(RAM) for training

We have 80000 images to train converted into 28x28 pixels images. We will list all files and assign them lables such as '0' for close eye and '1' for open eye

```python
import os
from os import listdir
import matplotlib as plt
import numpy as np
import tensorflow as tf

path='mrlEyes_2018_01/'
close_eyes_files=listdir(path+'close/')
open_eyes_files=listdir(path+'open/')
close_eyes_length=len(close_eyes_files)
open_eyes_length=len(open_eyes_files)
file_names=[]
print('close eye images count : ',close_eyes_length)
print('open eye images count : ',open_eyes_length)
length=min(close_eyes_length,open_eyes_length)

labels=[]

for i in range(length):
    file_names.append(path+'close/'+close_eyes_files[i])
    labels.append(int(0))
    file_names.append(path+'open/'+open_eyes_files[i])
    labels.append(int(1))
    
for i in range(length,close_eyes_length):
    file_names.append(path+'close/'+close_eyes_files[i])
    labels.append(int(0))
    
for i in range(length,open_eyes_length):
    file_names.append(path+'open/'+open_eyes_files[i])
    labels.append(int(1))
    
print('file_names length : ',len(file_names))
print('labels length : ',len(labels))

labels = tf.reshape(labels, (-1))
```

All listed files are loaded into RAM (2-4 GB approximately)
And convert 3 channel RGB images into Grayscale Single channel images.

```python
from matplotlib import image
from matplotlib import pyplot
import numpy as np
import tensorflow as tf
from tqdm import trange
import cv2
length=len(file_names)
dataset=[]
for i in trange(length):
    # load image as pixel array
    img=image.imread(fname=file_names[i])
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    dataset.append(img)
```

Dimensions of image array

```python
dataset=np.array(dataset)
dataset=dataset.reshape(-1,28,28,1)

dataset.shape
```

## Images loaded successfully

Display first 100 images to test

```python
class_names = ['close','open']
pyplot.figure(figsize=(15,15))
for i in range(100):
    pyplot.subplot(10,10,i+1)
    pyplot.xticks([])
    pyplot.yticks([])
    pyplot.grid(False)
    pyplot.imshow(dataset[i].reshape([28,28]))
    pyplot.xlabel(class_names[labels[i]])
pyplot.show()
```

```python
dataset = tf.reshape(dataset, (-1, 28, 28, 1))
```

## Build model

import all libraries for model building

```python
from matplotlib import image
import matplotlib.pyplot as plt
import numpy as np

import os
import PIL
import tensorflow as tf

from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
```

Split images array into two arrays:
1. train_images : first 50,000 images for model training
2. test_images : Remaing images used for testing accurecy

```python
train_size=50000
train_images = dataset[:train_size]
test_images = dataset[train_size:]
train_labels = labels[:train_size]
test_labels = labels[train_size:]
```

```python
print('train_images images : ',train_images.shape)
print('test_images images : ',test_images.shape)
print('train_labels images : ',train_labels.shape)
print('test_labels images : ',test_labels.shape)
```

Normalize pixel values between 0 and 1

```python
train_images= tf.cast(train_images, tf.float32) / 255.0
test_images= tf.cast(test_images, tf.float32) / 255.0
```

Data augmentation : rotate images randomly to increase accurecy

```python
data_augmentation = tf.keras.Sequential([
  layers.experimental.preprocessing.RandomFlip("horizontal_and_vertical"),
  layers.experimental.preprocessing.RandomRotation(0.2),
])
train_images=data_augmentation(train_images)
test_images=data_augmentation(test_images)
```

Create model

```python
model = Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.Flatten())
model.add(layers.Dense(64, activation='sigmoid'))
model.add(layers.Dense(2))
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

Model details

```python
model.summary()
```

Train model

```python
history = model.fit(train_images,train_labels,validation_data=(test_images,test_labels), epochs=5)
```

Accurecy of model by epochs

```python
pyplot.plot(history.history['accuracy'], label='accuracy')
pyplot.plot(history.history['val_accuracy'], label = 'val_accuracy')
pyplot.xlabel('Epoch')
pyplot.ylabel('Accuracy')
pyplot.ylim([0.5, 1])
pyplot.legend(loc='lower right')

test_loss, test_acc = model.evaluate(test_images,test_labels, verbose=2)
```

```python
print('Accurecy of model : ',test_acc)
```

Save model into 'h5' format so it can be loaded for prediction

```python
model.save('open_close_eyes_model.h5')
```
